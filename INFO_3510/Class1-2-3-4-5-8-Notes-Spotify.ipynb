```python
import requests
import os # os allows us to interact with the computer's operating system, including files

from dotenv import load_dotenv

load_dotenv("3510.env")
```

```python
clientID = os.getenv("SPOTIFY_CLIENT_ID")
```

```python
clientID
```

```python
clientSecret = os.getenv("SPOTIFY_CLIENT_SECRET")
```

```python
# authorization docs
# https://developer.spotify.com/documentation/web-api/concepts/authorization
```

```python
grantType = 'client_credentials'

bodyParams = {'grant_type':grantType}
url = "https://accounts.spotify.com/api/token"
```

```python
responseJSON = requests.post(url, data=bodyParams, auth = (clientID,clientSecret)).json()
```

```python
responseJSON
```

```python
responseJSON["access_token"]
```

```python
# spotipy
import spotipy
```

```python
# credentials
from spotipy.oauth2 import SpotifyClientCredentials
# spotipy keeps track of our access token
clientCredentialManager = SpotifyClientCredentials(client_id=os.getenv("SPOTIFY_CLIENT_ID"),client_secret=os.getenv("SPOTIFY_CLIENT_SECRET"))
```

```python
type(clientCredentialManager)
```

```python
# create spotipy object
sp = spotipy.Spotify(client_credentials_manager=clientCredentialManager)
```

```python
type(sp)
```

```python
# search for an artist
sp.search(q="Howling Giant",type="artist")
```

```python
for artist in sp.search(q="Howling Giant",type="artist")["artists"]["items"]:
    print(artist["name"])
```

```python
# recommendations
hgSearch = sp.search(q="Howling Giant",type="artist")
```

```python
hgSearch
```

```python
hgID =hgSearch["artists"]["items"][0]["id"]
hgID
```

```python
hgGlassFutureSearch = sp.search(q="Glass Future",type="track")
```

```python
hgGlassFutureSearch
```

```python
hgGlassFutureSearch["tracks"]["items"][0]
```

```python
# get glass future's spotify ID
gfID = hgGlassFutureSearch["tracks"]["items"][0]["id"]
```

```python
gfID
```

```python
# make recommendations based on artist ("Howling Giant") and track ("Glass Future")
hgRecs = sp.recommendations(seed_artists=[hgID],seed_tracks=[gfID],limit=5)
```
