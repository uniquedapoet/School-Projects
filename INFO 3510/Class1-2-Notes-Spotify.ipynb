```python
# set up spotify API access

# installed python-dotenv from command line using pip
```

```python
import requests
import os # os allows us to interact with the computer's operating system, including files

from dotenv import load_dotenv

load_dotenv("3510.env")
```

```python
clientID = os.getenv("SPOTIFY_CLIENT_ID")
```

```python
clientID
```

```python
clientSecret = os.getenv("SPOTIFY_CLIENT_SECRET")
```

```python
# authorization docs
# https://developer.spotify.com/documentation/web-api/concepts/authorization
```

```python
grantType = 'client_credentials'

bodyParams = {'grant_type':grantType}
url = "https://accounts.spotify.com/api/token"
```

```python
responseJSON = requests.post(url, data=bodyParams, auth = (clientID,clientSecret)).json()
```

```python
responseJSON
```

```python
responseJSON["access_token"]
```

```python
rootURL = 'https://api.spotify.com/v1/'
```

```python
headers = {"Authorization":"Bearer "+responseJSON["access_token"]}
```

```python
mastodonSearch = requests.get(rootURL+'search?q=Mastodon&type=artist',headers=headers).json()
```
