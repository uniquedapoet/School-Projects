```python
import requests
import os 
from dotenv import load_dotenv
import pprint as p

load_dotenv('3510.env')
```

```python
clientID = os.getenv('SPOTIPY_CLIENT_ID')
print(clientID)
```

```python
clientSecret = os.getenv('SPOTIFY_CLIENT_SECRET')
print(clientSecret)
```

```python
grantType = 'client_credentials'

bodyParams = {'grant_type' : grantType}

url = 'https://accounts.spotify.com/api/token'
```

```python
resposeJSON = requests.post(url, data=bodyParams, auth=(clientID, clientSecret)).json()
```

```python
p.pprint(resposeJSON)
```

```python
headers = {'Authorization' : 'Bearer ' + resposeJSON['access_token']}
```

```python
search_endpoint = 'https://api.spotify.com/v1/search'
search_params = {
    'q': 'mastodon',
    'type': 'artist',
    'limit': 10
}

response = requests.get(search_endpoint, params=search_params, headers=headers)
search_results = response.json()

# Print the search results
print(search_results)
```

```python
mastodonSearch = search_results['artists']['items'][0]['id']
```

```python
print(mastodonSearch)
rootURL = 'https://api.spotify.com/v1/'
```

```python
mastadonAlbums = requests.get(rootURL+'artists/'+mastodonSearch+'/albums', headers=headers).json()
```
